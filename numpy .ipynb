```python
#Array creation method
import numpy as np

# 1. np.array() ‚Äì Create array from list
arr1 = np.array([1, 2, 3, 4, 5])
print("np.array():\n", arr1)

# 2. np.zeros() ‚Äì Array of zeros
arr2 = np.zeros((3, 3),dtype=int)
print("\nnp.zeros():\n", arr2)

# 3. np.ones() ‚Äì Array of ones
arr3 = np.ones((2, 4),dtype=int)
print("\nnp.ones():\n", arr3)

# 4. np.full() ‚Äì Array with constant value
arr4 = np.full((3, 3), 7)
print("\nnp.full():\n", arr4)

# 5. np.eye() ‚Äì Identity matrix
arr5 = np.eye((4),dtype=int)
print("\nnp.eye():\n", arr5)

# 6. np.arange() ‚Äì Range of values
arr6 = np.arange(10, 51, 5)
print("\nnp.arange():\n", arr6)

# 7. np.linspace() ‚Äì Evenly spaced values
arr7 = np.linspace(0, 1, 6)
print("\nnp.linspace():\n", arr7)

# 8. np.random.rand() ‚Äì Random floats (0 to 1)
arr8 = np.random.rand(3, 3)
print("\nnp.random.rand():\n", arr8)

# 9. np.random.randint() ‚Äì Random integers
arr9 = np.random.randint(1, 100, 5)
print("\nnp.random.randint():\n", arr9)

# 10. np.empty() ‚Äì Uninitialized array
arr10 = np.empty((2, 2))
print("\nnp.empty():\n", arr10)
```

```python
#3. Modifying Using Indexing & Slicing
arr = np.array([1, 2, 3, 4, 5])
arr[2] = 99
print(arr)

mat = np.array([[1,2,3],[4,5,6]])
mat[:, 1] = 0
print(mat)
```

```python
#üîπ 4. Boolean Indexing ‚≠ê (Very Important)
#Condition-Based Selection
arr = np.array([10, 5, 20, 3, 15])

print(arr[arr > 10])

#Modify Using Condition
arr[arr < 10] = 0
print(arr)
```

```python
#6. Copy vs View (Exam Favorite)
#View (Changes affect original)
a = np.array([1, 2, 3, 4])
b = a[1:3]
b[:] = 99

print(a)
```

```python
#Copy (Independent array)
a = np.array([1, 2, 3, 4])
b = a[1:3].copy()
b[:] = 99

print(a)
```

```python
#reshaping & Manipulation
#1. reshape()
#Changes the shape of an array without changing data.

import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6])
reshaped = arr.reshape(2, 3)
print(reshaped)
#‚ö† Total elements must match
```

```python
#2. flatten()
#Converts a multi-dimensional array into a 1D array (copy).

mat = np.array([[1, 2],
                [3, 4]])

flat = mat.flatten()
print(flat)
#‚úî Returns a copy
```

```python
#3. ravel()
#Flattens array into 1D view (if possible).
mat = np.array([[1, 2],
                [3, 4]])
ravel_arr = mat.ravel()
print(ravel_arr)
#‚úî Faster than flatten()
#‚úî Usually returns a view
```

```python
#4. transpose() / .T
#Swaps rows and columns.
mat = np.array([[1, 2],
                [3, 4]])
print(mat.T)
```

```python
#5. resize()
#Changes shape in-place.
arr = np.array([1, 2, 3, 4])
arr.resize(2, 2)
print(arr)
#‚ö† Modifies original array
```

```python
#6. expand_dims()
#Adds a new axis (dimension).

arr = np.array([1, 2, 3])

print(np.expand_dims(arr, axis=0))  # Row
print(np.expand_dims(arr, axis=1))  # Column
```

```python
#7. squeeze()
#Removes dimensions of size 1.
arr = np.array([[[1, 2, 3]]])
print(arr.squeeze())
```

```python
#8. concatenate()
#Joins arrays along an axis.

a = np.array([[1, 2]])
b = np.array([[3, 4]])

print(np.concatenate((a, b), axis=0))
```

```python
#9. stack()
print(np.stack((a, b)))
#Other Stack Variants
print(np.vstack((a, b)))   # Vertical
print(np.hstack((a, b))) # Horizontal
```

```python
#10. split()
#Splits an array into multiple parts.

arr = np.array([1, 2, 3, 4, 5, 6])

print(np.split(arr, 3))
```
