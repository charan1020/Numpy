```python
#array
import numpy as np
list=np.array(['A','B','C','D'])
print(list)
```

```python
#integer
import numpy as np
list=np.array([1,2,3,4,5,6,7])
print(list[-1])
print(list[-2])
print(list[-3])
```

```python
#float
import numpy as np
list=np.array([1.2,4.5,3.4,66.7])
print(list[-1])
```

```python
#complex
import numpy as np
a=np.array([5+3j])
b=np.array([5+7j])
print(a+b)
```

```python
#boolean
import numpy as np
a=np.array([24])
print(bool(a))
```

# numpy methods

```python
#Append
import numpy as np
a=np.array([1,2,3,4])
b=np.append(a,5)
print(b)
```

```python
#insert
import numpy as np
list1=np.array(["mobile","instagram","reels"])
list2=np.insert(list1,1,"app")
print(list2)
```

```python
#delete
import numpy as np
a=np.array(["town","hospital","library"])
b=np.delete(a,2)
print(b)
```

```python
#copy
import numpy as np
a=np.array([1,2,3,4,5])
b=np.copy(a)
print(b)
```

```python
#sort
import numpy as np
a=np.array([4,5,6,1,2,3,7,8])
b=np.sort(a)
print(b)
```

```python
#reverse
import numpy as np
a=np.array([9,8,7,6,5,4,3,2,1])
b=a[::-1]
print(b)
```

```python
#unique
import numpy as np
a=np.array([1,2,3,1,2,3,4,3,2,3,3,2,3,7,8])
b=np.unique(a)
print(b)
```

```python
#size
import numpy as np
a=np.array([1,2,3,4])
b=np.size(a)
print(b)
```

```python
#clear
import numpy as np
a=np.array([1,2,3,4,5])
b=np.array([])
print(b)
```

```python
#SUM, MEAN, MAX, MIN
arr = np.array([1, 2, 3, 4, 5])
print("Sum:", arr.sum())
print("Mean:", arr.mean())
print("Max:", arr.max())
print("Min:", arr.min())
```

```python
#extend
import numpy as np
a=np.array([1,2,3,4])
b=np.array([5,6,7,8])
c=np.append(a,b)
print(c)
```
